```python
# Calcula o equilíbrio da economia

# Importa os módulos necessários
import numpy as np
import matplotlib.pyplot as plt

# Define os parâmetros
c_0 = 0.1
c_1 = 0.5
i_0 = 0.3
i_1 = 0.8
l_0 = 0.2
l_1 = 0.5
T = 15
G = 40
M = 62
P = 10

# Cria o grid
intervalo_r = np.linspace(0, 50, 1000)

# Cria a relação IS
IS = lambda r: ((c_0 - c_1 * T + i_0 - i_1 * r + G) / (1 - c_1))

# Cria a relação LM
LM = lambda r: (M + l_1 * P * r) / (P * l_0)

# Cria o excesso
funcao_excesso = lambda r: abs(IS(r) - LM(r))

# Cria a lista com os excessos
lista_excesso = []
for r in intervalo_r:
    lista_excesso.append(funcao_excesso(r)) # Lista dos excessos

# Calcula o excesso mínimo e o r e o Y de equilíbrio
excesso_minimo = np.argmin(lista_excesso) # Excesso mínimo
r_equilibrio = intervalo_r[excesso_minimo] # r de equilíbrio
Y_equilibrio = IS(r_equilibrio) # Y de equilíbrio

# Plotando o gráfico

# Cria as listas vazias
lista_IS = []
lista_LM = []

# Cria as curvas IS e LM
for r in intervalo_r:
    lista_IS.append(IS(r))
    lista_LM.append(LM(r))
    
# Outro modo de fazer
# lista_IS = IS(intervalo_r)
# lista_LM = LM(intervalo_r)

# Cria as curvas
fig, ax = plt.subplots(dpi = 500) # Nitidez
ax.plot(lista_IS, intervalo_r, label = "Curva IS",
        color = "green") # Relação IS
ax.plot(lista_LM, intervalo_r, label = "Curva LM",
        color = "red") # Relação LM

# Cria o ponto de equilíbrio e coloca a legenda
ax.plot(Y_equilibrio, r_equilibrio, markersize = 5,
        marker = "o", color = "black", label = "Ponto de equilíbrio") # Equilíbrio dos mercados
ax.legend(bbox_to_anchor=(0.001, 1.13, 0.5, 1.2), loc=3,
          fancybox=True, shadow = False, ncol = 3, borderaxespad=0) # Box de legendaplt.title("Modelo IS-LM")
ax.set_title("Modelo IS-LM")

# Define os intervalos e coloca a legenda nos eixos e no ponto
ax.set_xlim(0, Y_equilibrio * 2) # Intervalo do eixo x
ax.set_ylim(0, r_equilibrio * 2) # Intervalo do eixo y
ax.set_xlabel("Produto, Y") # Da nome ao eixo x
ax.set_ylabel("Taxa de juros, r") # Da nome ao eixo y
ax.annotate("%.2f" %(r_equilibrio),
            (1.05 * Y_equilibrio, 0.90 * r_equilibrio)) # Legenda o ponto de equilíbrio

# Coloca grade e plota o gráfico
plt.grid(True) # Coloca grade no gráfico
plt.show() # Plota o gráfico
```
